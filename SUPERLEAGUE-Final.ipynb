```python
def player_single(name1,name2,wl,i):  #wl=1,1 win;wl=2,2 win,i = ith round
    name1.r[i-1] = 1
    name2.r[i-1] = 1
    if name1.score >= name2.score:
        ad = name1   #st表示强弱，1=1强，2=2强
        da = name2
        st = 1
    elif name1.score < name2.score:
        ad = name2
        da = name1
        st = 2
    if wl == 1 and st == 1 :   #1胜1强
        (ad.dif,da.dif) = score_single(ad,da,i,1) 
    elif wl == 1 and st == 2:  #1胜2强
        (ad.dif,da.dif) = score_single(ad,da,i,2)  
    elif wl == 2 and st == 1 :  #2胜1强
        (ad.dif,da.dif) = score_single(ad,da,i,2)
    elif wl == 2 and st == 2:  #2胜2强
        (ad.dif,da.dif) = score_single(ad,da,i,1)
    if name1.score >= name2.score:
        name1.dif = ad.dif
        name2.dif = da.dif
    elif name1.score < name2.score:
        name2.dif = ad.dif
        name1.dif = da.dif
    refresh_score([name1,name2],i)
```

```python
def player_double(name1,name2,name3,name4,wl,i):  #wl=1,1 win;wl=2,2 win,i = ith round
    name1.r[i-1] = 1
    name2.r[i-1] = 1
    name3.r[i-1] = 1
    name4.r[i-1] = 1
    if name1.score + name2.score >= name3.score + name4.score:
        #ad = qa   #st表示强弱，1=1强，2=2强
        #da = ru
        qa.score = (name1.score + name2.score)/2  
        ru.score = (name3.score + name4.score)/2
        st = 1
    elif  name1.score + name2.score < name3.score + name4.score:
        #ad = qa   #st表示强弱，1=1强，2=2强
        #da = ru
        qa.score = (name3.score + name4.score)/2  
        ru.score = (name1.score + name2.score)/2
        st = 2
    if wl == 1 and st == 1 :   #1胜1强
        (qa.dif,ru.dif) = score_single(qa,ru,i,1)
    elif wl == 1 and st == 2:  #1胜2强
        (qa.dif,ru.dif) = score_single(qa,ru,i,2)
    elif wl == 2 and st == 1 :  #2胜1强
        (qa.dif,ru.dif) = score_single(qa,ru,i,2)
    elif wl == 2 and st == 2:
        (qa.dif,ru.dif) = score_single(qa,ru,i,1)
    if  name1.score + name2.score >= name3.score + name4.score:
        name1.dif = qa.dif/4.00
        name2.dif = qa.dif/4.00
        name3.dif = ru.dif/4.00
        name4.dif = ru.dif/4.00
    elif name1.score + name2.score < name3.score + name4.score:
        name1.dif = ru.dif/4.00
        name2.dif = ru.dif/4.00
        name3.dif = qa.dif/4.00
        name4.dif = qa.dif/4.00
    refresh_score([name1,name2,name3,name4],i)
```

```python
def score_single(qiang,ruo,i,st):  #who=1强者胜，=2弱者胜
    qiang.dif = 0
    ruo.dif = 0
    if st == 1:
        if 0 <= qiang.score - ruo.score < 1:
            qiang.dif = qiang.dif + 10
            ruo.dif = ruo.dif - 5
        elif 1 <= qiang.score - ruo.score < 26:
            qiang.dif = qiang.dif + 9
            ruo.dif = ruo.dif - 4.5
        elif 26 <= qiang.score - ruo.score < 51:
            qiang.dif = qiang.dif + 8
            ruo.dif = ruo.dif - 4
        elif 51 <= qiang.score - ruo.score < 101:
            qiang.dif = qiang.dif + 7
            ruo.dif = ruo.dif - 3.5
        elif 101 <= qiang.score - ruo.score < 151:
            qiang.dif = qiang.dif + 6
            ruo.dif = ruo.dif - 3
        elif 151 <= qiang.score - ruo.score < 201:
            qiang.dif = qiang.dif + 5
            ruo.dif = ruo.dif - 2.5
        elif 201 <= qiang.score - ruo.score < 301:
            qiang.dif = qiang.dif + 4
            ruo.dif = ruo.dif - 2
        elif 301 <= qiang.score - ruo.score < 401:
            qiang.dif = qiang.dif + 3
            ruo.dif = ruo.dif - 1.5
        elif 401 <= qiang.score - ruo.score < 501:
            qiang.dif = qiang.dif + 2
            ruo.dif = ruo.dif - 1
        elif 501 <= qiang.score - ruo.score < 701:
            qiang.dif = qiang.dif + 1
            ruo.dif = ruo.dif - 0.5
        elif qiang.score - ruo.score >= 701:
            qiang.dif = qiang.dif + 0
            ruo.dif = ruo.dif - 0
    elif st == 2:
        if 0 <= qiang.score - ruo.score < 1:
            qiang.dif = qiang.dif - 5
            ruo.dif = ruo.dif + 10
        elif 1 <= qiang.score - ruo.score < 26:
            qiang.dif = qiang.dif - 5.5
            ruo.dif = ruo.dif + 11
        elif 26 <= qiang.score - ruo.score < 51:
            qiang.dif = qiang.dif - 6
            ruo.dif = ruo.dif + 12
        elif 51 <= qiang.score - ruo.score < 101:
            qiang.dif = qiang.dif - 7
            ruo.dif = ruo.dif + 14
        elif 101 <= qiang.score - ruo.score < 151:
            qiang.dif = qiang.dif - 8
            ruo.dif = ruo.dif + 16
        elif 151 <= qiang.score - ruo.score < 201:
            qiang.dif = qiang.dif - 10
            ruo.dif = ruo.dif + 20
        elif 201 <= qiang.score - ruo.score < 301:
            qiang.dif = qiang.dif - 12
            ruo.dif = ruo.dif + 24
        elif 301 <= qiang.score - ruo.score < 401:
            qiang.dif = qiang.dif - 14
            ruo.dif = ruo.dif + 28
        elif 401 <= qiang.score - ruo.score < 501:
            qiang.dif = qiang.dif - 16
            ruo.dif = ruo.dif + 32
        elif 501 <= qiang.score - ruo.score < 701:
            qiang.dif = qiang.dif - 18
            ruo.dif = ruo.dif + 36
        elif qiang.score - ruo.score >= 701:
            qiang.dif = qiang.dif - 20
            ruo.dif = ruo.dif + 40
    return qiang.dif,ruo.dif
```

```python
def refresh_score(li,i):
    for l in li:
        pr = 0.00
        pr = sum(l.r)/i
        weigh = weight(pr)
        l.score  = l.score + l.dif
        l.real_score = weigh * l.score
```

```python
def weight(p):
    if 0 <= p <= 0.1:
        p = 3*p
    elif 0.1 < p <= 0.3:
        p = 0.5 * p + 0.25
    elif 0.3 < p <= 0.5:
        p = 2 * p - 0.2
    elif 0.5 < p <= 0.8:
        p = 0.5 * p + 0.55
    elif 0.8 < p <= 1:
        p = 0.25 * p +0.75
    return p
```

```python
def printresult(li):
    cname =[]
    cscore = []
    crealscore = []
    for l in li:
        cname = cname + [l.name]
        cscore = cscore + [l.score]
        crealscore = crealscore + [l.real_score]
        d = {'权衡实力对比积分' : pd.Series(crealscore, index=cname),'实力对比积分' : pd.Series(cscore, index=cname),}
        d1 = pd.DataFrame(d, index=cname)
    return(d1)
```
